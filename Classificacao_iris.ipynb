# Flor de Íris

Iris é a família da flor que contém as várias espécies como a iris setosa, iris versicolor, iris virginica, etc.

## 1. Criar o dataset

Para classificar as diferentes espécies da íris, devemos preparar os conjuntos de dados com características e rótulos. Mas o sklearn vem com os conjuntos de dados embutidos para o problema de classificação da íris.

Vamos primeiro entender os conjuntos de dados

O conjunto de dados consiste em:

* 150 amostras
* 3 rótulos: espécies de Iris (Iris setosa, Iris virginica e Iris versicolor)
* 4 características: comprimento da sépala, largura da sépala, comprimento da pétala, largura da pétala em cm



O Scikit learn só funciona se os dados forem armazenados como dados numéricos, independentemente de ser uma regressão ou um problema de classificação. Também requer que os arrays sejam armazenados em arrays numpy para otimização. Como esse conjunto de dados é carregado do scikit learn, tudo é formatado adequadamente.

Então agora vamos escrever o código python para carregar o conjunto de dados Iris.

```python
from sklearn import  datasets

iris = datasets.load_iris()
iris
```

Atribua os dados e o destino a variáveis separadas.

```python
x = iris.data
y = iris.target
```

```python
x
```

x contém os atributos e y contém os rótulos

## Dividindo o conjunto de dados


Como nosso processo envolve treinamento e teste, devemos dividir nosso conjunto de dados. Ele pode ser executado pelo seguinte código

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x, y, test_size=.75)
```

x_train contém os atributos de treinamento

x_test contém os atributos de teste

y_train contém o rótulo (classe) de treinamento

y_test contém os rótulos (classe) de teste

## 2. Construa o modelo

Podemos usar qualquer algoritmo de classificação para resolver o problema. Nós resolvemos o problema anterior com o algoritmo da árvore de decisão, eu vou com isso.

```python
from sklearn import tree

classifier=tree.DecisionTreeClassifier()
```

O código acima criará o modelo vazio. Para fornecer as operações ao modelo devemos treiná-los.

> Nota:Também podemos usar KNeighborsClassifier (a eficiência é maior)

```python
from sklearn import neighbors

classifier = neighbors.KNeighborsClassifier()
```

Neste ponto, acabamos de fazer o modelo. Mas ele não pode prever se uma determinada flor pertence a qual espécie de íris. Se nosso modelo tem que prever a flor, temos que treinar o modelo com os atributos e os rótulos.

## 3.Treine o Modelo.

Podemos treinar o modelo com função de ajuste (fit).

```python
classifier.fit(x_train,y_train)
```

Agora o modelo está pronto para fazer previsões

## 4. Faça previsões:

As previsões podem ser feitas com a função de previsão (predict).

```python
predictions = classifier.predict(x_test)
```

essas previsões podem ser combinadas com a saída esperada para medir o valor de acurácia.

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_test, predictions))
```
